Valid Python notebooks that still exist on github

```python
with connect() as session:
    sql = """
        SELECT r.id, n.id, r.repository, r.commit, n.name, r.notebooks_count, r.notebooks
        FROM repositories r, notebooks n
        WHERE r.id = n.repository_id
        AND n.skip & (2048 + 1024 + 512 + 128) = 0
        AND n.processed & 15 = 0
        AND n.language = 'python'
    """
    df = pd.read_sql(sql, session.connection())
    
```

I replaced the sampling function to load from a existing file to prevent resampling

```python
#sample = df.sample(69)
sample = pd.read_excel("samples.xls")
```

```python
for i, (_, result) in enumerate(sample.iterrows()):
    name = result.repository.split("/")[-1]
    print()
    print("git clone https://github.com/{}.git".format(result.repository))
    print("cd {}".format(name))
    print("git checkout {}".format(result.commit))
    print("cd ..")
    print("mv {} {:02}.{}".format(name, i + 1, name))
```

```python
sample.to_excel("data/samples.xls")
```
