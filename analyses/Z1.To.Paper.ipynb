```python
import shutil
from pathlib import Path

 
paperpath = Path("../paper")
outputs = Path("outputs")

from analysis_helpers import load_vars, var

def check_sample(mode, temp):
    if mode == "d_" and '%' in current[temp] and not temp.endswith("_lift"):
        sd_temp = "sd_" + temp[2:]
        if sd_temp in variables:
            v_d = float(variables[temp].replace('\\%', ''))
            v_sd = float(variables[sd_temp].replace('\\%', ''))
            
            d_sd_percentages[temp] = (v_d, v_sd)
            d_sd_total[temp] = (int(variables[temp+"_total"].replace(',', '')), int(variables[sd_temp+"_total"]))

def use(
    name, modes=["d_"], paper=[], 
    preattempt=["", "times_", "timesone_"], 
    posattempt=["", "_total"], warn=True
):
    for mode in modes:
        for pre in preattempt:
            for pos in posattempt:
                temp = mode + pre + name + pos
                if temp in variables:
                    current[temp] = variables[temp]
                    check_sample(mode, temp)
                elif warn and pos == "" and pre == "":
                    print(temp, "not found")
                    
def use_file(name, paper=[], crop=False):
    target = str(paperpath / "figures" / name) 
    source = str(outputs / name)
    if crop:
        !pdfcrop $source $target
    else:
        shutil.copy(str(source), str(target))

def use_manual_file(name, paper=[]):
    pass

def load():
    variables = load_vars()
    variables = {
        k: v for k, v in variables.items()
        if not k.startswith("sd_diff")
        if not k.startswith("sd_sup")
        if not k.startswith("td_diff")
        if not k.startswith("td_sup")
    }
    return variables
```
